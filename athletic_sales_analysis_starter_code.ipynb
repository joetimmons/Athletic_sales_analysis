```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
df_2020 = pd.read_csv('Resources/athletic_sales_2020.csv')
df_2021 = pd.read_csv('Resources/athletic_sales_2021.csv')

# Display the first few rows of the 2020 sales DataFrame
df_2020.head()

```

```python
# Display the 2020 sales DataFrame
df_2020
```

```python
# Display the 2021 sales DataFrame
df_2021
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
df_2020.dtypes
```

```python
# Check the 2021 sales data types.
df_2021.dtypes
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
combined_df = pd.concat([df_2020, df_2021], axis=0).reset_index(drop=True)

combined_df
```

```python
# Check if any values are null.
null_values = combined_df.isnull().any()

null_values
```

```python
# Check the data type of each column
column_data_types = combined_df.dtypes

column_data_types
```

```python
# Convert the "invoice_date" to a datetime datatype
combined_df['invoice_date'] = pd.to_datetime(combined_df['invoice_date'], format='%Y-%m-%d')

# Check the data types to confirm the conversion
combined_df.dtypes
```

```python
# Confirm that the "invoice_date" data type has been changed.
invoice_date_data_type = combined_df['invoice_date'].dtypes

invoice_date_data_type


```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
print(combined_df.columns)
```

```python
# Group by region, state, and city, and sum the products sold
products_sold = combined_df.groupby(['region', 'state', 'city'])['units_sold'].sum().reset_index()

# Rename the summed column to "Total_Products_Sold"
products_sold.rename(columns={'units_sold': 'Total_Products_Sold'}, inplace=True)

# Show the top 5 results
top_5_products_sold = products_sold.nlargest(5, 'Total_Products_Sold')

top_5_products_sold


                            
```

#### Using `pivot_table`

```python
# Create a pivot table
pivot_table = combined_df.pivot_table(
    values='units_sold', 
    index=['region', 'state', 'city'], 
    aggfunc='sum'
)

# Rename the aggregated column to 'Total_Products_Sold'
pivot_table.rename(columns={'units_sold': 'Total_Products_Sold'}, inplace=True)

# Show the pivot table
pivot_table
```

```python
# Display the top 5 results
top_5_products_sold = pivot_table.sort_values('Total_Products_Sold', ascending=False).head(5)

top_5_products_sold

```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with 'invoice_date' as the index and 'total_sales' as the values
pivot_table = combined_df.pivot_table(
    values='total_sales', 
    index='invoice_date', 
    aggfunc='sum'
)

# Optionally, rename the aggregated column to 'Total Sales'
pivot_table.columns = ['Total Sales']

# Show the pivot table
pivot_table

```

```python
# First, ensure that 'invoice_date' is in datetime format
combined_df['invoice_date'] = pd.to_datetime(combined_df['invoice_date'])

# Create the initial pivot table
pivot_table = combined_df.pivot_table(
    values='total_sales',
    index='invoice_date',
    aggfunc='sum'
)

# Resample the pivot table into daily bins and sum the total sales for each day
daily_sales = pivot_table.resample('D').sum()

# Rename the column to 'Total Sales'
daily_sales.columns = ['Total Sales']

# Sort the resampled pivot table in ascending order based on 'Total Sales'
sorted_daily_sales = daily_sales.sort_values(by='Total Sales', ascending=True)

sorted_daily_sales

```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# First, ensure that 'invoice_date' is in datetime format
combined_df['invoice_date'] = pd.to_datetime(combined_df['invoice_date'])

# Create the initial pivot table
pivot_table = combined_df.pivot_table(
    values='total_sales',
    index='invoice_date',
    aggfunc='sum'
)

# Resample the pivot table into weekly bins and sum the total sales for each week
weekly_sales = pivot_table.resample('W').sum()

# Rename the column to 'Total Sales'
weekly_sales.columns = ['Total Sales']

# Sort the resampled pivot table in ascending order based on 'Total Sales'
sorted_weekly_sales = weekly_sales.sort_values(by='Total Sales', ascending=True)

sorted_weekly_sales

```
